```python
# Import Libraries
import numpy as np
import pandas as pd
```

```python
# load and convert files into DataFrames
freedom_to_load = "../Resources/freedom.csv"
happiness_to_load = "../Resources/happiness.csv"
income_to_load = "../Resources/income.csv"
suicide_to_load = "../Resources/suicide.csv"

freedom_data = pd.read_csv(freedom_to_load)
happiness_data = pd.read_csv(happiness_to_load)
income_data = pd.read_csv(income_to_load)
suicide_data = pd.read_csv(suicide_to_load)
```

```python
# clean up freedom csv, first by filtering for 2015 data
# metrics for freedom indices are scaled out of 10
freedom_data = freedom_data.loc[freedom_data["year"]==2015]
freedom_data
```

```python
# delete unwanted columns from DataFrame
freedom_df = pd.DataFrame({
    "Country":freedom_data["countries"],
    "Personal Freedom":freedom_data["pf_score"],
    "Economic Freedom":freedom_data["ef_score"],
    "Human Freedom":freedom_data["hf_score"]
})

# drop rows with NaN in them
freedom_df = freedom_df.dropna(axis=0,how="any")
freedom_df
```

```python
# view columns to see which ones to pick out
happiness_data.columns
```

```python
# clean up happiness csv, first by filtering for 2015 data
happiness_data = happiness_data.loc[happiness_data["year"]==2015]

happiness_df = pd.DataFrame({
    "Country":happiness_data["country"],
    "Happiness":happiness_data["Life Ladder"],
    "Generosity":happiness_data["Generosity"],
    "Perception of Corruption":happiness_data["Perceptions of corruption"],
    "Confidence in Government":happiness_data["Confidence in national government"],
    "Gini Index":happiness_data["GINI index (World Bank estimate), average 2000-13"],
})

# drop rows with NaN in them
happiness_df = happiness_df.dropna(axis=0,how="any")
happiness_df
```

```python
# clean up income csv
income_data
```

```python
# only useful information is the income group, so filter for that column 
income_df = pd.DataFrame({
    "Country":income_data["ShortName"],
    "Income Group":income_data["IncomeGroup"]
})
income_df = income_df.dropna(axis=0,how="any")
income_df
```

```python
# clean up suicide csv
# filtered to data in 2015 because 2016 data for most countries is incomplete
suicide_data.set_index("country")
suicide_data = suicide_data.loc[suicide_data["year"]==2015]
suicide_data
```

```python
suicide_data["suicides_no"] = pd.to_numeric(suicide_data["suicides_no"])
suicide_data["population"] = pd.to_numeric(suicide_data["population"])
suicide_filtered = suicide_data[[
    "country",
    "sex",
    "suicides_no",
    "population"
]]
```

```python
# count number of suicides by country and sex
suicide_count = suicide_filtered.groupby(by=["country","sex"])["suicides_no"].sum()
suicide_count = pd.DataFrame(suicide_count)
suicide_count
```

```python
# sum up population country and country
pop_count = suicide_filtered.groupby(by=["country","sex"])["population"].sum()
pop_count = pd.DataFrame(pop_count)
pop_count
```

```python
# merge both data frames
suicide_comb = pd.merge(suicide_count, pop_count, on=["country","sex"])
suicide_comb
```

```python
# find number of suicides per 100k people
suicides_per_100k = []
for index,row in suicide_comb.iterrows():
    per_100k = row["suicides_no"]*100000/row["population"]
    suicides_per_100k.append(per_100k)
suicides_per_100k
```

```python
# split suicides per 100k people list into two lists
i=0
per_male = []
per_female = []
for i in range(0,len(suicides_per_100k)-1,2):
    per_female.append(suicides_per_100k[i])
    per_male.append(suicides_per_100k[i+1])
print(per_female)
print(per_male)
```

```python
# make list of Total GDP and GDP per capita
total_gdp = []
gdp_capita = []
gdp_list = suicide_data[" gdp_for_year ($) "].tolist()
capita_list = suicide_data["gdp_per_capita ($)"].tolist()
i=0

# gdp numbers change every 12 rows
for i in range(0,len(gdp_list),12):
    total_gdp.append(gdp_list[i])
gdp_list

for i in range(0,len(capita_list),12):
    gdp_capita.append(capita_list[i])
```

```python
# create new DataFrame for suicide data
suicide_df = pd.DataFrame({
    "Country":suicide_data["country"].unique(),
    "Suicides per 100k Males":per_male,
    "Suicides per 100k Females":per_female,
    "Total GDP":total_gdp,
    "GDP per Capita":gdp_capita
})
suicide_df
```

```python
combined_data = pd.merge(happiness_df, freedom_df, on="Country", how="inner")
combined_data = pd.merge(combined_data, income_df, on="Country", how="inner")
combined_data = pd.merge(combined_data, suicide_df, on="Country", how="inner")
combined_data
```

```python
# export data to csv file
combined_data.to_csv("")
```
